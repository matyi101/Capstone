```r
library("readxl")
library("plotly")
require(ggplot2)

# Load Crude Oil Pricing Data:
# Daily closing price of Crude Oil (WTI, BRENT AND BONNY)

CrudePrice <- read_excel("E:/Google Drive/690 - Capstone Project/04 Working Directory (Codes)/CrudePrices.xlsx")
GasPrice <- read_excel("E:/Google Drive/690 - Capstone Project/04 Working Directory (Codes)/GasPrices.xlsx")
DieselPrice <- read_excel("E:/Google Drive/690 - Capstone Project/04 Working Directory (Codes)/DieselPrices.xlsx")
```

```r
summary (CrudePrice)
summary (GasPrice)
summary (DieselPrice)
```

```r
colors <- c("WTI" = "blue", "BRENT" = "red", "BONNY" = "steelblue")

plot1 <- ggplot(CrudePrice, aes(Date, color = factor(season))) +                   
  geom_line(aes(y=WTI, colour="WTI"), size = 0.75) + 
  geom_line(aes(y=BRENT, colour="BRENT"), size = 0.75) +  
  geom_line(aes(y=BONNY, color="BONNY"), size = 0.75) + 
  ggtitle("Crude Oil Curve") + theme(plot.title = element_text(hjust = 0.5)) + 
  labs(x = " ",
         y = "Price (USD/bbl)",
         color = "Legend") +
    scale_color_manual(values = colors)


plot1

```

```r
x <- GasPrice$Date
USGas <- GasPrice$US_Gasoline
UKGas <- GasPrice$UK_Gasoline
NGGas <- GasPrice$Nigeria_Gasoline

colors <- c("USGas" = "blue", "UKGas" = "red", "NGGas" = "orange")
plot2 <- ggplot(GasPrice, aes(x)) +                    
  geom_line(aes(y=USGas, colour="USGas"), size = 0.75) + 
  geom_line(aes(y=UKGas, colour="UKGas"), size = 0.75) +  
  geom_line(aes(y=NGGas, color="NGGas"), size = 0.75) +
  ggtitle("Plot of Gasoline Price Curve") + theme(plot.title = element_text(hjust = 0.5)) +
  labs(x = " ",
         y = "Price (USD/gal)",
         color = "Legend") +
    scale_color_manual(values = colors)
plot2
```

```r
y <- DieselPrice$Date
US <- DieselPrice$US_Diesel
UK <- DieselPrice$UK_Diesel
NG <- DieselPrice$Nigeria_Diesel

colors <- c("US" = "blue", "UK" = "red", "NG" = "orange")
plot3 <- ggplot(DieselPrice, aes(y)) +                    
  geom_line(aes(y=US, colour="US"), size = 0.75) + 
  geom_line(aes(y=UK, colour="UK"), size = 0.75) +  
  geom_line(aes(y=NG, color="NG"), size = 0.75) +
  ggtitle("Plot of Diesel Price Curve") + theme(plot.title = element_text(hjust = 0.5)) +
  labs(x = " ",
         y = "Price (USD/gal)",
         color = "Legend") +
    scale_color_manual(values = colors)
plot3
```

```r
library(grid)
# Move to a new page
grid.newpage()
# Create layout : nrow = 3, ncol = 2
pushViewport(viewport(layout = grid.layout(nrow = 3, ncol = 2)))
# A helper function to define a region on the layout
define_region <- function(row, col){
  viewport(layout.pos.row = row, layout.pos.col = col)
} 
# Arrange the plots
print(plot1, vp = define_region(row = 1, col = 1:2))   # Span over two columns
print(plot2, vp = define_region(row = 2, col = 1))
print(plot3, vp = define_region(row = 2, col = 2))
```
