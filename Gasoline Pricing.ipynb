```python
# EXPLORATORY DATA ANALYSIS - BASIC STATISTIC

# Libraries
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
sns.set()
```

```python
Gas = pd.read_excel('GasPrices_bbl.xlsx', engine='openpyxl')

USGas = Gas.US_Gasoline
UKGas = Gas.UK_Gasoline
NGGas = Gas.NG_Gasoline
date = Gas.Date
```

```python
Gas.describe()
```

```python
Gas.skew() 
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
sns.lineplot(x=date, y=USGas)
plt.ylim(0,200)
plt.ylabel('Price/bbl')
plt.title('US Gasoline')

plt.subplot(1, 3, 2)
sns.lineplot(x=date, y=UKGas, color="teal")
plt.ylim(0,300)
plt.ylabel('Price/bbl')
plt.title('UK Gasoline')

plt.subplot(1, 3, 3)
sns.lineplot(x=date, y=NGGas, color="purple")
plt.ylim(0,200)
plt.ylabel('Price/bbl')
plt.title('NG Gasoline')

plt.show()

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
USGas.hist(bins = 25)
plt.title('US Gasoline')

plt.subplot(1, 3, 2)
UKGas.hist(bins = 25, color='teal')
plt.title('UK Gasoline')

plt.subplot(1, 3, 3)
NGGas.hist(bins = 25, color='purple')
plt.title('NG Gasoline')

plt.show()

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
sns.kdeplot(USGas, shade=True, color="r")
plt.title('WTI')

plt.subplot(1, 3, 2)
sns.kdeplot(UKGas, shade=True, color="b")
plt.title('BRENT')

plt.subplot(1, 3, 3)
sns.kdeplot(NGGas, shade=True, color="g")
plt.title('BONNY LIGHT')

plt.show()

# Gaussian Distribution Plot


plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
x_values = np.arange(25, 200, 0.01)
y_values = sp.stats.norm(USGas.mean(), USGas.std())
plt.plot(x_values, y_values.pdf(x_values))

plt.subplot(1, 3, 2)
x_values = np.arange(175, 325, 0.01)
y_values = sp.stats.norm(UKGas.mean(), UKGas.std())
plt.plot(x_values, y_values.pdf(x_values))

plt.subplot(1, 3, 3)
x_values = np.arange(0, 100, 0.01)
y_values = sp.stats.norm(NGGas.mean(), NGGas.std())
plt.plot(x_values, y_values.pdf(x_values))

plt.show()
```
