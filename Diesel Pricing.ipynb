```python
# EXPLORATORY DATA ANALYSIS - DIESEL

# Libraries
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
sns.set()
```

```python
Diesel = pd.read_excel("DieselPrices_bbl.xlsx", engine='openpyxl')
USDiesel = Diesel.US_Diesel
UKDiesel = Diesel.UK_Diesel
NGDiesel = Diesel.Nigeria_Diesel
date = Diesel.Date
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
sns.lineplot(x=date,y=USDiesel)
plt.ylim(40,325)
plt.ylabel('Price/bbl')
plt.title('US Diesel')

plt.subplot(1, 3, 2)
sns.lineplot(x=date, y=UKDiesel, color="teal")
plt.ylim(40,325)
plt.ylabel('Price/bbl')
plt.title('UK Diesel')

plt.subplot(1, 3, 3)
sns.lineplot(x=date, y=NGDiesel, color="purple")
plt.ylim(40,325)
plt.ylabel('Price/bbl')
plt.title('NG Diesel')

#Histogram

plt.show()

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
USDiesel.hist(bins = 50)

plt.subplot(1, 3, 2)
UKDiesel.hist(bins = 50, color='teal')

plt.subplot(1, 3, 3)
NGDiesel.hist(bins = 50, color='purple')


#Density Plot

plt.show()

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
sns.kdeplot(USDiesel, shade=True, color="r")
plt.title('WTI')

plt.subplot(1, 3, 2)
sns.kdeplot(UKDiesel, shade=True, color="b")
plt.title('BRENT')

plt.subplot(1, 3, 3)
sns.kdeplot(NGDiesel, shade=True, color="g")
plt.title('BONNY LIGHT')

plt.show()

# Gaussian Distribution Plot


plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
x_values = np.arange(50, 175, 0.01)
y_values = sp.stats.norm(USDiesel.mean(), USDiesel.std())
plt.plot(x_values, y_values.pdf(x_values))

plt.subplot(1, 3, 2)
x_values = np.arange(175, 325, 0.01)
y_values = sp.stats.norm(UKDiesel.mean(), UKDiesel.std())
plt.plot(x_values, y_values.pdf(x_values))

plt.subplot(1, 3, 3)
x_values = np.arange(25, 135, 0.01)
y_values = sp.stats.norm(NGDiesel.mean(), NGDiesel.std())
plt.plot(x_values, y_values.pdf(x_values))

plt.show()
```
